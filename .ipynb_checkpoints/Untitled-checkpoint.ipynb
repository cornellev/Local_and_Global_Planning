```python
import numpy as np
from scipy.optimize import minimize

def inflection_constraints(coefficients, n):
    inflection_points = 0
    for i in range(2, n):
        if np.polyder(coefficients, i).shape[0] > 2:
            inflection_points += len(np.roots(np.polyder(coefficients, i)))
    return inflection_points - n + 2

def objective(coefficients, waypoints):
    x = waypoints[:, 0]
    y = waypoints[:, 1]
    p = np.polyval(coefficients, x)
    return np.sum((p - y) ** 2)

def fit_polynomial(waypoints, degree):
    initial_guess = np.zeros(degree + 1)
    result = minimize(objective, initial_guess, args=(waypoints,), method='Nelder-Mead')
    return result.x

# Example waypoints (x, y)
waypoints = np.array([
    [0, 0],
    [1, 1],
    [2, -1],
    [3, 0]
])

# Degree of the polynomial (n - 1)
degree = len(waypoints) - 1

# Fit polynomial
coefficients = fit_polynomial(waypoints, degree)

# Ensure inflection points
constraints_result = minimize(inflection_constraints, coefficients, args=(degree,), method='Nelder-Mead')

if constraints_result.success:
    optimized_coefficients = constraints_result.x
    print("Optimized Coefficients:", optimized_coefficients)
else:
    print("Optimization failed.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def inflection_constraints(coefficients, n, inflection_point_index):
    inflection_points = 0
    for i in range(2, n):
        if np.polyder(coefficients, i).shape[0] > 2:
            inflection_points += len(np.roots(np.polyder(coefficients, i)))
    return inflection_points - n + 2 + np.polyder(coefficients, 2)[inflection_point_index]

def objective(coefficients, waypoints):
    x = waypoints[:, 0]
    y = waypoints[:, 1]
    p = np.polyval(coefficients, x)
    return np.sum((p - y) ** 2)

def fit_polynomial(waypoints, degree, inflection_point_index):
    initial_guess = np.zeros(degree + 1)
    result = minimize(objective, initial_guess, args=(waypoints,), method='Nelder-Mead')
    coefficients = result.x
    # Ensure inflection points
    constraints_result = minimize(inflection_constraints, coefficients, args=(degree, inflection_point_index), method='Nelder-Mead')
    if constraints_result.success:
        return constraints_result.x
    else:
        print("Optimization failed.")
        return coefficients

# Example waypoints (x, y)
waypoints = np.array([
    [0, 0],
    [1, 1],
    [2, -1],
    [3, 0]
])

# Index of the waypoint where we want the inflection point
inflection_point_index = 1  # Change this index as needed

# Degree of the polynomial (n - 1)
degree = len(waypoints) - 1

# Fit polynomial
coefficients = fit_polynomial(waypoints, degree, inflection_point_index)

# Generate points for plotting the polynomial curve
x_values = np.linspace(np.min(waypoints[:, 0]), np.max(waypoints[:, 0]), 100)
y_values = np.polyval(coefficients, x_values)

# Plot the waypoints and the polynomial curve
plt.figure(figsize=(8, 6))
plt.plot(waypoints[:, 0], waypoints[:, 1], 'bo', label='Waypoints')
plt.plot(x_values, y_values, 'r-', label='Optimized Polynomial Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Curve Fitting with Waypoints and Inflection Constraint')
plt.legend()
plt.grid(True)
plt.show()
```
