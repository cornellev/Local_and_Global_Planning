```python
import numpy as np

def bezier_curve(control_points, n_points=100):
    t = np.linspace(0, 1, n_points)
    n = len(control_points)
    curve = np.zeros((n_points, 2), dtype=np.float64)  # Ensure float64 dtype

    def bernstein_poly(i, n, t):
        return np.math.comb(n - 1, i) * t**i * (1 - t)**(n - 1 - i)

    for i in range(n):
        curve += np.column_stack([bernstein_poly(i, n, t), bernstein_poly(i, n, t)]) * control_points[i]

    return curve

# Example usage:
for x in range(len(waypoints)):
    waypoints[x] = np.float64(x)
    
control_points = np.array(waypoints, dtype=np.float64)  # Ensure float64 dtype
curve_points = bezier_curve(control_points)

import matplotlib.pyplot as plt

plt.plot(curve_points[:, 0], curve_points[:, 1], label='Bezier Curve')
plt.scatter(control_points[:, 0], control_points[:, 1], c='red', label='Control Points')
plt.legend()
plt.show()
```

```python
import numpy as np

def bezier_curve(control_points, num_points=100):
    t = np.linspace(0, 1, num_points)
    n = len(control_points) - 1
    result = np.zeros((num_points, 2))

    for i in range(num_points):
        point = np.zeros(2)
        for j, p in enumerate(control_points):
            point += p * np.math.factorial(n) / (np.math.factorial(j) * np.math.factorial(n - j)) * \
                     (1 - t[i]) ** (n - j) * t[i] ** j
        result[i] = point

    return result

# Example usage:
waypoints = [[5, 5], [19, 19], [33, 33], [55, 55], [53, 74], [51, 94], [49, 114], [47, 134], [46, 154], [44, 174], [42, 194], [40, 214], [38, 234], [37, 254], [35, 274], [33, 294], [31, 313], [29, 333], (340, 340), [28, 353], [362, 356], [378, 368], [26, 373], [394, 381], [413, 386], [432, 392], [24, 393], [451, 398], [470, 404], [489, 410], [22, 413], [508, 416], [527, 422], [547, 428], [20, 433], [19, 453], [553, 457], [557, 477], [16, 489], [561, 496], [20, 508], [565, 516], [25, 527], [569, 535], [29, 547], [573, 555], [34, 566], [577, 575], [38, 586], [43, 605], [577, 613], [577, 633], [52, 644], [69, 653], [577, 653], [87, 662], [105, 671], [577, 673], [123, 680], [141, 689], [577, 693], [158, 698], [176, 707], [577, 713], [194, 716], [212, 725], [577, 733], [230, 734], [247, 743], [265, 753], [578, 753], [283, 762], [543, 764], [301, 771], [525, 771], [506, 778], [319, 780], [487, 784], [337, 789], [468, 791], [354, 798], [449, 798], [430, 804], [372, 807], [411, 811], [393, 818]]
control_points = np.array(waypoints)
curve_points = bezier_curve(control_points)

# Plotting the Bzier curve
import matplotlib.pyplot as plt

plt.plot(control_points[:, 0], control_points[:, 1], 'ro--', label='Control Points')
plt.plot(curve_points[:, 0], curve_points[:, 1], 'b-', label='Bzier Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bzier Curve')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
