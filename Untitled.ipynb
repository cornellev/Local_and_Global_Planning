```python
import numpy as np
from scipy.optimize import minimize

def inflection_constraints(coefficients, n):
    inflection_points = 0
    for i in range(2, n):
        if np.polyder(coefficients, i).shape[0] > 2:
            inflection_points += len(np.roots(np.polyder(coefficients, i)))
    return inflection_points - n + 2

def objective(coefficients, waypoints):
    x = waypoints[:, 0]
    y = waypoints[:, 1]
    p = np.polyval(coefficients, x)
    return np.sum((p - y) ** 2)

def fit_polynomial(waypoints, degree):
    initial_guess = np.zeros(degree + 1)
    result = minimize(objective, initial_guess, args=(waypoints,), method='Nelder-Mead')
    return result.x

# Example waypoints (x, y)
# waypoints = np.array([
#     [0, 0],
#     [1, 1],
#     [2, -1],
#     [3, 0]
# ])

waypoints = np.array([(5, 5), (55, 55), (37, 592), (53, 629), (167, 688), (465, 790), (593, 688), (544, 413), (399, 391), (340, 340)])

# Degree of the polynomial (n - 1)
degree = len(waypoints) - 1

# Fit polynomial
coefficients = fit_polynomial(waypoints, degree)

# Ensure inflection points
constraints_result = minimize(inflection_constraints, coefficients, args=(degree,), method='Nelder-Mead')

if constraints_result.success:
    optimized_coefficients = constraints_result.x
    print("Optimized Coefficients:", optimized_coefficients)
else:
    print("Optimization failed.")
```
